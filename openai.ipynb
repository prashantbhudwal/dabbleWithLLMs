```python
from rich import print
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=api_key)

response = client.chat.completions.create(
    messages=[{
        "role": "user",
        "content": "Write a story on dogs and cats in 2 sentences.",
    }],
    model="gpt-4o-mini",
)

print(response)
```

```python
sample_messages = [
    {
        "role": "system",
        "content": "You are a helpful assistant."
    },
    {
        "role": "user",
        "content": "Who dog is golden, friendly, and fluffy? One sentence answer."   
    },
]   
```

```python
response = client.chat.completions.create(    
    messages = sample_messages,
    model ="gpt-4o-mini",
    temperature = 0.7,
    logprobs=True,
)
print(response.choices[0].message.content)
print(response.choices[0].logprobs)

    
```

```python
def has_low_confidence(choice, threshold):
    # First access the logprobs object from the choice
    logprobs_obj = choice.logprobs

    # Now iterate over the token-level objects in the logprobs content
    for token_obj in logprobs_obj.content:
        logprob = token_obj.logprob
        token_str = token_obj.token
        if logprob is None:
            continue
        if logprob < threshold:
            print(f"Token '{token_str}' has low confidence: {logprob}")
            return True
    return False
```

```python
LOW_CONFIDENCE_THRESHOLD = -3.0
has_low_confidence(response.choices[0], LOW_CONFIDENCE_THRESHOLD)
```
